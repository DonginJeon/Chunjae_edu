```python
from transformers import pipeline
from PIL import Image
import matplotlib.pyplot as plt
```

```python
pipe = pipeline(
    "document-question-answering", model="naver-clova-ix/donut-base-finetuned-docvqa"
)
```

```python
document = Image.open("./dataset/travel.jpeg")
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(document)
plt.axis(False)
plt.show()
```

```python
question = "what is price of cheese?"
answer = pipe(image=document, question=question)
answer
```

연산이 가능한 함수

```python
def total_pay():
    price = 0
    product = input("상품이름을 입력하세요, 여러개일 경우 ,로 구분해주세요")
    product = product.split(",")

    for pro in product:
        question = f"what is price of {pro}?"
        answer = pipe(image=document, question=question)
        a = answer[0].get("answer")
        a = float(a)
        price += a

    return price
```

데코레이터를 활용하는 것으로 개선해보자

```python
total_pay()
```
