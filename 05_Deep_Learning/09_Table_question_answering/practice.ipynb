```python
import folium
from geopy.geocoders import Nominatim

# 주소를 입력합니다.
address = "강원도 강릉시 지변동 123 강릉원주대학교"

# geopy를 사용하여 주소를 좌표로 변환합니다.
geolocator = Nominatim(user_agent="geoapiExercises")
location = geolocator.geocode(address)

# 좌표 확인
latitude = location.latitude
longitude = location.longitude

# 지도 생성 (초기 위치를 해당 좌표로 설정)
mymap = folium.Map(location=[latitude, longitude], zoom_start=15)

# 마커 추가
folium.Marker([latitude, longitude], popup="강릉원주대학교").add_to(mymap)

# 지도를 HTML 파일로 저장
mymap.save("gangneung_university_map.html")

# 지도 띄우기 (주피터 노트북 환경에서)
mymap
```

```python
table = {"columns": list(df.columns), "data": df.values.tolist()}

# 질문 정의
query = "What are the addresses in 강원도?"

# 모델 호출
result = qa_model(table=table, query=query)

# 결과 출력
addresses = result["answer"]
print(f"Addresses in 강원도: {addresses}")
```

```python
def movie_quotes():
    movie = input("알고 싶은 영화제목을 입력하세요")
    if movie in df["Movie Name"].unique():
        question = qa_model(query=f"{movie}에 나온 quote를 알려줘", table=df)
        return question
    else:
        return "영화정보를 찾을 수 없습니다"
```
