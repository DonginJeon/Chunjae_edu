```python
import os
from moviepy.editor import VideoFileClip

# 비디오 파일이 있는 폴더 경로
video_folder_path = "./new_video/"

# 오디오 파일을 저장할 폴더 경로
audio_output_folder_path = "./new_audio/"

# 오디오 파일을 저장할 폴더가 없으면 생성
if not os.path.exists(audio_output_folder_path):
    os.makedirs(audio_output_folder_path)

# 비디오 폴더 내 모든 파일을 가져오기
for filename in os.listdir(video_folder_path):
    # 확장자가 비디오 파일인지 확인 (여기서는 .mp4 기준)
    if filename.endswith(".mp4"):
        video_path = os.path.join(video_folder_path, filename)
        audio_output_path = os.path.join(
            audio_output_folder_path, f"{os.path.splitext(filename)[0]}.mp3"
        )

        # 비디오 파일 불러오기
        video_clip = VideoFileClip(video_path)

        # 비디오에서 오디오 추출
        audio_clip = video_clip.audio

        # 오디오 파일로 저장
        print(f"Converting: {filename} -> {audio_output_path}")
        audio_clip.write_audiofile(audio_output_path)

        # 리소스 해제
        audio_clip.close()
        video_clip.close()

print("All video files have been converted to audio!")
```

```python
from transformers import pipeline

whisper_small = pipeline("automatic-speech-recognition", "openai/whisper-small")
```

```python
from moviepy.editor import AudioFileClip
import os


# 오디오 파일을 30초 단위로 나누는 함수
def split_audio(file_path, chunk_length=30):
    audio = AudioFileClip(file_path)
    total_duration = int(audio.duration)
    audio_chunks = []

    for start in range(0, total_duration, chunk_length):
        end = min(start + chunk_length, total_duration)
        chunk = audio.subclip(start, end)
        chunk_path = f"chunk_{start}_{end}.mp3"
        chunk.write_audiofile(chunk_path)
        audio_chunks.append(chunk_path)

    return audio_chunks


# 각 청크에서 텍스트 추출
def transcribe_audio_chunks(chunk_files):
    transcriptions = []
    for chunk_file in chunk_files:
        result = whisper_small(chunk_file, generate_kwargs={"language": "Korean"})
        transcriptions.append(result["text"])
        os.remove(chunk_file)  # 임시 파일 삭제
    return transcriptions


# 전체 텍스트를 하나로 합침
def merge_transcriptions(transcriptions):
    return " ".join(transcriptions)


# 오디오 파일 경로
audio_file = audio_output_folder_path


# 오디오 파일을 30초 단위로 나누고 텍스트 추출
chunks = split_audio(audio_file)
transcriptions = transcribe_audio_chunks(chunks)
final_transcription = merge_transcriptions(transcriptions)
```

```python
import os

# 오디오 파일이 저장될 폴더 경로
audio_output_path = "./new_audio/"
# 텍스트 파일을 저장할 폴더 경로
text_output_path = "./new_text/"

# 텍스트 파일을 저장할 폴더가 없으면 생성
if not os.path.exists(text_output_path):
    os.makedirs(text_output_path)

# 오디오 파일들을 처리하는 반복문
for audio_file in os.listdir(audio_output_path):
    if audio_file.endswith(".mp3") or audio_file.endswith(
        ".wav"
    ):  # 오디오 파일 확장자 필터링
        audio_file_path = os.path.join(audio_output_path, audio_file)

        # 오디오 파일을 30초 단위로 나누고 텍스트 추출
        chunks = split_audio(audio_file_path)
        transcriptions = transcribe_audio_chunks(chunks)
        final_transcription = merge_transcriptions(transcriptions)

        # 결과 텍스트 파일로 저장
        text_file_name = f"{os.path.splitext(audio_file)[0]}.txt"
        text_file_path = os.path.join(text_output_path, text_file_name)

        with open(text_file_path, "w", encoding="utf-8") as text_file:
            text_file.write(final_transcription)

        print(f"Saved transcription to {text_file_path}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding

# 데이터 로딩
data = df_k

# labels_dict
labels_dict = {
    "수": 0,
    "덧셈": 1,
    "뺄셈": 2,
    "곱셈": 3,
    "나눗셈": 4,
    "분수": 5,
    "소수": 6,
    "소수점": 7,
    "길이": 8,
    "시간": 9,
    "화폐": 10,
    "측정": 11,
    "각도": 12,
    "평면도형": 13,
    "입체도형": 14,
    "도형": 15,
    "대칭": 16,
    "원": 17,
    "직각": 18,
    "비율": 19,
    "비례식": 20,
    "비": 21,
    "그래프": 22,
    "통계": 23,
    "확률": 24,
    "약수": 25,
    "배수": 26,
    "함수": 27,
    "미지수": 28,
    "규칙": 29,
}

# Reverse dictionary for label to text conversion
reverse_labels_dict = {v: k for k, v in labels_dict.items()}

# 데이터 전처리
texts = data["text"].values
tags1 = data["태그1"].fillna(-1).astype(int)  # NaN 값을 -1로 처리하고 정수형으로 변환
tags2 = data["태그2"].fillna(-1).astype(int)
tags3 = data["태그3"].fillna(-1).astype(int)

# Tokenization
tokenizer = Tokenizer(num_words=100)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
X = pad_sequences(sequences, maxlen=100)


# Encoding tags
def encode_tags(tags):
    tags_encoded = np.zeros((len(tags), len(labels_dict)))  # One-hot encoded tags
    for i, tag in enumerate(tags):
        if tag >= 0 and tag < len(labels_dict):
            tags_encoded[i, tag] = 1
    return tags_encoded


# 태그를 리스트로 변환하여 인코딩
y1 = encode_tags(tags1.tolist())
y2 = encode_tags(tags2.tolist())
y3 = encode_tags(tags3.tolist())

# Train-test split
X_train, X_test, y1_train, y1_test, y2_train, y2_test, y3_train, y3_test = (
    train_test_split(X, y1, y2, y3, test_size=0.2, random_state=42)
)

# 모델 정의
model = Sequential()
model.add(Embedding(input_dim=100, output_dim=30, input_length=100))
model.add(LSTM(64, return_sequences=True))
model.add(LSTM(32))
model.add(Dense(len(labels_dict), activation="sigmoid"))

# 컴파일
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# 학습
model.fit(
    X_train, y1_train, epochs=10, batch_size=32, validation_data=(X_test, y1_test)
)


# 예측 함수
def predict_tags(text):
    sequence = tokenizer.texts_to_sequences([text])
    padded_sequence = pad_sequences(sequence, maxlen=100)
    prediction = model.predict(padded_sequence)
    return [reverse_labels_dict[idx] for idx in np.where(prediction[0] > 0.5)[0]]


# 새로운 텍스트 예측
new_text = """당신이 보는 개는 몇 마리입니까? 여기에 개들이 있습니다. 우리는 그들을 세어야만 합니다. 그래서 이 개부터 세기 시작하는데, 그를 다시 세지 않도록 주의해야 합니다. 그래서 저는 가장 위에 있는 이 개부터 시작하겠다고 기억하고 있습니다. 하나, 둘, 셋, 넷, 다섯, 여섯, 일곱, 여덟. 제가 본 개는 여덟 마리입니다. 아홉 마리로 가려고 할 수도 있지만, 이미 이 개를 세었습니다. 그래서 개는 단지 여덟 마리뿐입니다. 그래서 여덟 마리의 개를 봤습니다. 그럼 우리의 끝을 확인해 봅시다. 잘 했습니다. 당신이 보는 쥐는 몇 마리입니까? 똑같습니다. 저는 가장 윗쪽에 있는 쥐부터 세기 시작하는 것을 좋아하지만, 가장 위의 쥐를 계속 세고 있으면 안된다는 것을 기억해야 합니다. 그래서 하나, 둘, 셋, 넷, 다섯, 여섯, 일곱. 이제 여기에 일곱 마리의 쥐가 있는데, 아마 여덟 마리로 가려고 할 수도 있겠지만, 우리는 이미 이 위의 쥐를 세었습니다. 그래서 여기에는 쥐가 단지 일곱 마리뿐입니다. 일곱 마리 쥐. 당신이 동그랗게 계속 돌면서 쥐를 반복해서 세는 것은 조심해야 합니다. 당신은 각각의... 쥐를 딱 한 번만 세고 싶을 것입니다. 다른 것도 해 봅시다. 이것은 너무 재미 있습니다. 
당신이 보는 쿠키는 몇 개입니까? 자, 모두 여기에 섞여있네요. 그리고 우리는 쿠키를 두 번 세지 않도록 조심해야만 합니다. 그럼 보겠습니다. 하나, 둘, 셋, 넷, 다섯, 여섯, 일곱, 여덟, 아홉. 쿠키는 아홉 개입니다. 저는 아홉 개의 쿠키를 봅니다. 쥐는 몇 마리입니까? 자, 이제 여기에는 쥐가 많이 있군요. 그러니까 하나, 둘, 셋, 넷, 다섯, 여섯, 일곱. 여덟, 아홉, 열, 열하나, 열둘, 열셋, 열넷, 열다섯, 열여섯, 열일곱, 열여덟, 열아홉 마리의 쥐입니다. 쥐는 열아홉 마리입니다. 쥐가 많네요."""
predicted_tags = predict_tags(new_text)
print(predicted_tags)
```
