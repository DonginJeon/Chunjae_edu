```python
# 데이터베이스에 연결
conn = sqlite3.connect("example.db")
cursor = conn.cursor()

# 테이블 생성 (영상 데이터를 저장할 BLOB 필드 포함)
cursor.execute(
    """
CREATE TABLE IF NOT EXISTS video_files (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    video BLOB
)
"""
)


# 영상 파일을 읽어 BLOB 데이터로 변환
def read_video_file(file_path):
    with open(file_path, "rb") as file:
        return file.read()


# 영상 파일을 데이터베이스에 저장
file_path = "./khan_academy_kids/Video_ Comparing numbers of objects.mp4"
video_data = read_video_file(file_path)
cursor.execute(
    "INSERT INTO video_files (name, video) VALUES (?, ?)", ("video.mp4", video_data)
)

# 변경사항 저장 및 연결 종료
conn.commit()
conn.close()

print("영상 파일이 데이터베이스에 저장되었습니다.")
```
