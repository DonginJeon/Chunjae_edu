# CTT
- 초기에 사용된 교육측정 이론의 모델
- 학생의 검사 관찰 점수를 진점수와 오차의 합으로 가정
- 검사의 총점을 이용해 문항의 특성과 검사의 신뢰도 등을 파악

=> 간단한 절차를 바탕으로 문항 분석과 검사분석을 할 수 있다
## 단점

# IRT
- 문항마다 보편적인 고유한 특성을 가진다고 가정
    - 특정 집단에 한정X

- 문항 단위로 피험자에 대한 표준화된 분석이 가능?
- 적은 문항 수로 신뢰도 높은 검사 결과를 받아볼 수 있나
- 같은 문항을 풀지 않아도 같은 점수를 비교할 수 있나
- 표본 그룹과 독립적으로 판단가능한가

## 가정
- 일차원성 가정
    - 연속적인 일차원의 척도 체계를 가짐
- 증가성 가정
    - 학습자의 능력이 높을수록 맞출 확률이 높음
- 지역독립성 가정
    - 피험자가 맞출 확률, 서로 다른 피험자가 문항의 답을 맞힐 확률은 상호 독립적이다
- 불변성 가정
    - 어떤 검사를 이용하더라도 피험자의 능력 수준을 일관성있게 추정할 수 있다

### 문항반응함수 
- 피험자의 능력 수준에 따라 문항별 특성값을 바탕으로 문항 정답 확률을 계한하는 함수
    - 1PL(parameter Logistic) : 난이도
    - 2 :난이도 변별
    - 3 : 난이도 추측 변별

#### 모형
- 로지스틱 모형
    - 로지스틱 시그모이드 함수 :x값에 따라 y값을 0과 1 사이의 값으로 반환
- 정규 오자이브 모형

### 문항반응곡선
- 문항특성곡선 : 문항반응함수를 곡선으로 나타낸 것

### 문항모수
- 문항 변별도
    - 피험자를 변별하는 정도를 나타내는 지수
    - 문항특성곡선의 정답률이 0.5가 되는 위치의 기울기
- 문항 난이도
    - 문항의 쉽고 어려운 정도
    - 문항을 맞힐 확률이 0.5일떄 해당되는 능력 수준의 값
- 문항 추측도
    - 선다형 문항에서 문항을 추측하여 맞추는 비율을 나타내는 지수
    - 문항특성곡선에서 능력 수준이 음의 무한대로 접근할떄 수렴하는 값


### 검사특성곡선(TCC)
- 문항별 문항반응 함수를 합하여 검사의 총점을 문항별 1점인 검사 점수로 파악하는 것
- 검사 결과와 학생 능력 수준의 대응관계를 파악 가능

### 표준오차
- 개별 학생의 능력 수준 측정값이 얼마나 정확한지 나타내는 지표
- 학생의 능력 수준에 따른 검사정보함수 값의 역수의 양의 제곱근


### 문항반응이론과 적응형 검사
#### 컴퓨터 기반 적응형 검사(CAT)
- 학생의 능력 수준에 맞추어 실시간으로 개인 맞춤형으로 진행되는 컴퓨터 기반 검사
- 정오답 여부에 따라 학생을 기준으로 가장 큰 문항정보값을 갖는 문항을 제공한다.
- 표준오차의 길이를 바탕으로 신뢰구간을 측정

### 검사종류기준
- 표준오차 값이 일정 값 이하가 될때
- 추정되는 학생 능력 수준의 변화가 거의 없을때
- 시험 시간 만족
- 문제 수 만족

### 문제은행(DB)
#### 문제은행
- CAT를 위해서는 문항 모수에 대한 추정이 끝난 문항들이 담겨 있는 문제 은행이 필요하다
- 학습할때 정오답데이터를 막 넣지말고 노출도를 보면서 생각

### 최대우도법
- 주어진 함수 값을 최대화하는 변수의 값을 찾는 방법

#### EM 알고리즘(3PL)
- 순서1. 초기 문항 모수 및 학생 수준 설정
- 순서2. 학생에 대한 문항별 정답 확률 계산
- 순서3. 초기 문항 모수를 바탕으로 학생 능력 수준 추정
- 순서4. M-step: 추정된 학생 능력 수준을 바탕으로 문항모수 추정
