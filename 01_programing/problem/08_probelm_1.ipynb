````python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# 조건 설정을 위한 객체 호출
from selenium.webdriver.common.by import By


service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
genre_list = ["travel_2", "romance_8", "fiction_10"]
genre_name = ["travel", "romance", "fiction_"]
for genre in genre_list:
    url = f"https://books.toscrape.com/catalogue/category/books/{genre}/index.html"

    driver.implicitly_wait(1)

    # 웹 페이지 열기
    driver.get(url)
    information_list = driver.find_elements(
        By.CSS_SELECTOR, "li.col-xs-6.col-sm-4.col-md-3.col-lg-3"
    )

    for information in information_list:

        title1 = information.find_element(By.CSS_SELECTOR, "h3")
        title = title1.text
        price1 = information.find_element(By.CSS_SELECTOR, "p.price_color")
        price = price1.text
        info1 = information.find_element(By.CSS_SELECTOR, "a")
        info2 = info1.get_attribute("href")
        info = info2

        data_dict = {"카테고리": genre, "책 이름": title, "가격": price, "주소": info}
        print(data_dict)

    try:
        next_button2 = driver.find_element(By.PARTIAL_LINK_TEXT, "next")
        next_button2.click()
    except :
        break
    print(data_dict)


# quote_tags = driver.find_elements(By.CSS_SELECTOR, "div.tags")
# for tag_ in quote_tags:
#     tag_name = tag_.find_element(By.CSS_SELECTOR, "a.tag")
#     print(tag_name.text)
````
