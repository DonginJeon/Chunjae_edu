```python
import seaborn as sns
import matplotlib.pyplot as plt


titanic = sns.load_dataset("titanic")

print(titanic.head())
```

2. `age` 와 `fare` 의 관계를 Scatter Plot으로 시각화하세요.
    
    ![Untitled](https://prod-files-secure.s3.us-west-2.amazonaws.com/f2678325-6f7b-4a25-b188-86c42030d6d5/ba54cba1-0375-473d-8033-5594ed6e5c03/Untitled.png)

```python
sns.scatterplot(data=titanic, x="age", y="fare")
plt.show()
```

3. `age` 와 `fare` 의 관계를 `survived` 로 구분해서 Scatter Plot으로 시각화하세요.
    
    ![Untitled](https://prod-files-secure.s3.us-west-2.amazonaws.com/f2678325-6f7b-4a25-b188-86c42030d6d5/c2d48678-9efc-4b1e-8956-409d3c99fc64/Untitled.png)

```python
sns.scatterplot(data=titanic, x="age", y="fare", hue="survived")
plt.show()
```

```python
import seaborn as sns


diamonds = sns.load_dataset("diamonds")

print(diamonds.head())
```

2. `price` 와 `carat` 의 관계를 Scatter Plot으로 시각화하세요.
    
    ![Untitled](https://prod-files-secure.s3.us-west-2.amazonaws.com/f2678325-6f7b-4a25-b188-86c42030d6d5/4baf4d18-5bfb-4caf-876e-b06e4a787000/Untitled.png)

```python
sns.scatterplot(data=diamonds, x="price", y="carat")
plt.show()
```

3. `price` 와 `carat` 의 관계를 `color` 로 구분해서 Scatter Plot으로 시각화하세요.
    
    ![Untitled](https://prod-files-secure.s3.us-west-2.amazonaws.com/f2678325-6f7b-4a25-b188-86c42030d6d5/f1e4e39e-2fc2-42e5-9398-7c2724a04970/Untitled.png)

```python
sns.scatterplot(data=diamonds, x="price", y="carat", hue="color")
plt.show()
```

4. `price` 와 `carat` 의  관계를 `cut` 로 구분해서 Scatter Plot으로 시각화하세요.
    
    ![Untitled](https://prod-files-secure.s3.us-west-2.amazonaws.com/f2678325-6f7b-4a25-b188-86c42030d6d5/54797177-b334-4034-a7f1-bbf7bc1dbe19/Untitled.png)

```python
sns.scatterplot(data=diamonds, x="price", y="carat", hue="cut")
```

5. `price` 와 `carat` 의 관계를 `cut` 품질에 따라 구분해서 다중 시각화하세요.
    
    ![Untitled](https://prod-files-secure.s3.us-west-2.amazonaws.com/f2678325-6f7b-4a25-b188-86c42030d6d5/1505c2e9-28f8-4064-944c-92b40932a61c/Untitled.png)

```python
g = sns.FacetGrid(diamonds, col="cut")
g.map(sns.scatterplot, "carat", "price")
plt.show()
```
