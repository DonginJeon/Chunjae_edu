```python
import folium

# 초기 중심 위치와 줌 레벨을 설정한 지도 생성
m = folium.Map(location=[37.5665, 126.9780], zoom_start=50)

# 지도 저장
m.save("map.html")

# 노트북에 지도 출력
m
```

```python
# 마커 MARKER
"""folium.Marker(
    location,
    popup,
    tooltip,
    icon
)
"""

popup = folium.Popup(
    max_width=300,
    min_width= 100,
    html="<b>서울시청<b>"
)


icon = folium.Icon(
    icon="glass",
    color="skyblue",
)
folium.Marker(
    location=[37.5665, 126.9780],
    # popup = popup,
    icon=icon,
    tooltip="클릭"
    
).add_to(m) # 맵 객체에 마커를 추가
m
```

```python
# 두번째 마커 생성
import folium

m = folium.Map(location=[37.5665, 126.9780], zoom_start=13)


# 첫 번째 마커의 Popup 객체
name = "서울시청"
html = f"<p style='font-size:10px'>{name}</p>"
popup = folium.Popup(
    html=html,
    max_width=100,
)

# 아이콘 객체생성
icon = folium.Icon(icon="cloud", color="blue")

# 첫 번째 마커 생성 및 추가
folium.Marker(
    location=[37.5665, 126.9780],
    popup=popup,
    tooltip="클릭!",
    icon=icon,
).add_to(m)

m

```
