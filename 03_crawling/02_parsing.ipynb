```python
import requests

# beautiful soup 모듈 불러오기
from bs4 import BeautifulSoup

# 요청 주소
url = "https://example.com"

# 요청 - get(url) 메서드
response = requests.get(url)

# 상태코드속성
status_code = response.status_code

# 정상적인 요청이라면
if status_code == 200:

    # HTML 문서 속성,
    print(response.text)

    # beautifulsoup 객체 생성
    soup = BeautifulSoup(response.text, "html.parser")
    print(soup)

    # BeautifulSoup. find(태그이름) : 탐색 메서드
    p_tag = soup.find("p")
    print(p_tag)
    print(type(p_tag))

    # soup 인스턴스 내부에서 h1태그를 찾아서 택스트 출력
    h1_tag = soup.find("h1")
    print(h1_tag)
```

```python
import requests

# beautiful soup 모듈 불러오기
from bs4 import BeautifulSoup

# 요청 주소
url = "https://example.com"

# 요청 - get(url) 메서드
response = requests.get(url)

# 상태코드속성
status_code = response.status_code

# 정상적인 요청이라면
if status_code == 200:

    # # HTML 문서 속성,
    # print(response.text)

    # beautifulsoup 객체 생성
    soup = BeautifulSoup(response.text, "html.parser")

    # 실습
    a_tag_list = soup.find_all("a")
    for a_tag in a_tag_list:
        print(a_tag.text)

    # 태그의 속성 추출
    print(a_tag["href"])

    # 태그의 속성 목록 추출
    a_tag_attr_list = a_tag.attrs
    print(a_tag_attr_list)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://quotes.toscrape.com/"

response = requests.get(url)

# html 문서 파싱
soup = BeautifulSoup(response.text, "html.parser")

# div 태그, class가 quote인 요소 복수 탐색
quote_list = soup.find_all("div",attrs={"class":"quote"})

quote_author_list = []
for quote in quote_list:
    #quote 요소 내부에서 태그는 span 클래스는 text 요소 탐색
    quote_text = quote.find("span",attrs = {"class": "text"})
    # print(quote_text.text)

    quote_author = quote.find("small", attrs={"class":"author"})
    print(quote_author.text)
    # print(author_text.text)
    quote_author_dict = {
        "quote": quote_text.text,
        "author": quote_author.text
    }
    quote_author_list.append(quote_author_dict)

from pprint import pprint
pprint(quote_author_list)
```
