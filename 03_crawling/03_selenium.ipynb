```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# 조건 설정을 위한 객체 호출
from selenium.webdriver.common.by import By


service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
url = "https://quotes.toscrape.com"

driver.implicitly_wait(10)

# 웹 페이지 열기
driver.get(url)

quote_list = driver.find_elements(By.CLASS_NAME, "quote")
print(len(quote_list))

quotes = []
for quote in quote_list:
    quote_text = quote.find_element(By.CLASS_NAME, "text")
    quote_author = quote.find_element(By.CLASS_NAME, "author")
    print(f"{quote_text.text}-{quote_author.text}")
    quote_dict = {"quote": quote_text.text, "quthor": quote_author.text}
    quotes.append(quote_dict)

from pprint import pprint

pprint(quotes)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# 조건 설정을 위한 객체 호출
from selenium.webdriver.common.by import By


service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
url = "https://quotes.toscrape.com"

driver.implicitly_wait(10)

# 웹 페이지 열기
driver.get(url)
quotes = []
for i in range(10):

    quote_list = driver.find_elements(By.CLASS_NAME, "quote")

    for quote in quote_list:
        quote_text = quote.find_element(By.CLASS_NAME, "text")
        quote_author = quote.find_element(By.CLASS_NAME, "author")
        quote_dict = {"quote": quote_text.text, "quthor": quote_author.text}
        quotes.append(quote_dict)

    if i == 9:
        break
    next_button2 = driver.find_element(By.PARTIAL_LINK_TEXT, "Next")
    next_button2.click()

from pprint import pprint

pprint(quotes)
```
