### 패키지

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

plt.rcParams["font.family"] = "NanumGothic"
```

### 데이터호출

```python
df = pd.read_csv("./Data/Customed_df.csv")
```

### 사진 수에 따른 데이터 정리

```python
photo_1 = df[df["product_photos_qty"] == 1]

photo_2 = df[df["product_photos_qty"] == 2]


photo_3 = df[df["product_photos_qty"] == 3]


photo_4 = df[df["product_photos_qty"] == 4]

photo_5 = df[df["product_photos_qty"] == 5]


photo_6 = df[df["product_photos_qty"] == 6]

photo_7 = df[df["product_photos_qty"] == 7]


photo_8 = df[df["product_photos_qty"] == 8]


photo_9 = df[df["product_photos_qty"] == 9]

photo_10 = df[df["product_photos_qty"] == 10]
```

### 카테고리별 판매량에 대한 그래프

```python
data_v = pd.DataFrame(df["product_category_name"].value_counts())
x = df["product_category_name"].value_counts().index
y = df["product_category_name"].value_counts().values

px.bar(data_v, x=x, y=y)
```

```python
df["product_category_name"].unique()
```

### 사진 수에 따른 카테고리별 판매량 그래프

```python
photo = [
    photo_1,
    photo_2,
    photo_3,
    photo_4,
    photo_5,
    photo_6,
    photo_7,
    photo_8,
    photo_9,
    photo_10,
]


for p in photo:

    titles = [f"사진이 {i}장일 때 그래프" for i in range(1, 11)]
    datav = p["product_category_name"].value_counts().head(13)
    data_v = pd.DataFrame(datav)
    x = datav.index
    y = datav.values
    plt.figure(figsize=(30, 9))
    sns.barplot(data=data_v, x=x, y=y)
    plt.title(titles)
    plt.show()
```

### 상품 금액 분포에 대한 그래프

```python
top13_categories = photo_1["product_category_name"].value_counts().head(13).index

for i in top13_categories:

    plt.figure(figsize=(16, 6))

    d6 = pd.DataFrame(
        photo_1[photo_1["product_category_name"] == i].loc[:, ["price", "product_id"]]
    )

    sns.scatterplot(data=d6, x=d6["price"], y=d6.index, hue="product_id")

    plt.title(f"{i}의 분포")

    plt.legend([], [], frameon=False)
```

### 가설
- 해당 사이트에서 제품에 대한 상세한 설명이 제품 판매에 영향을 줄 것이다.

1. 사진의 수에 따라 판매량의 순위에는 영향을 안줄것이다. -> 실제 사진 수가 늘어남에 따라 순위 변동이 생김.
    - 사진이 많을수록 어떤 상품이 잘 팔릴까?
        - 사진이 많다, 제품을 꼼꼼히 살핀다.(예시)
        


2. 카테고리(침대, 목욕용품)를 특정하여 상품 금액 분포를 확인
    - 금액 분포에 큰 차이는 없지만 사진 수가 5장일 때 특정 금액으로 몰리는 경향이 있음 -> 인지도에 따른 선택일 것이다(몰렸다는 것은 같은 브랜드의 상품이고 보다 더 저렴한 상품일 것이다.)
    - 사진이 많을수록 금액의 범위가 넓어짐(1장일 땐 최고가가 800정도였는데 늘어날수록 2000정도로 넓어짐), 4장일 땐 600으로 범위가 좁아지고 상품 수도 줄어듦.
    - 비교적 저렴한 금액은 목욕용품이랑 식탁보 등 일상생활용품이고 가격이 비교적 높은 금액은 침대일 것이라 추정
        - 사진이 1장일 항목에는 침대 항목에 대한 수가 적을 것이라고 추측, 반면 사진이 늘수록 침대 항목이 추가되지 않을
    - 해당 카테고리는 1,2,3,5장일 땐 13위까지의 순위에 있는 반면 이후부터는 10장까지도 순위권에 없음. -> 즉 해당 카테고리를 구매할 때 사이트에서 고민하기보다는 애초에 제품 자체를 구매할 의사가 있어서 들어온 것으로 보임. 가장 많이 팔리는 카테고리인만큼 사진이 판매에 영향(마케팅적 차원)을 직접적으로 미친다고 말하기는 어렵울 거 같다.

3. 비싼 제품일수록 사진이 구매요소에 영향을 줄 것인가 -> 비싼 제품일수록 상세한 설명을 위해 사진을 많이 첨부할 것이다
    - 애초에 사이트 자체의 상품 금액의 분포가 500이하에 밀집되어 있고 소수의 고가 제품이 있기는 하나 이 사이트의 주된 타겟 상품이 아니라고 생각.
    - 사이트에서 다루는 비싼 제품의 기준이 필요함
        - 1500이라는 기준을 잡고 최고가가 1700정도인 장난감을 봤을 때 장난감은 사진의 수의 상관없이 순위권안에 항상들어오고 10위에선 가장 많은 판매량, 또한 유일하게 20장이 있는 카테고리도 장난감이다.
        - 장난감은 보통 부모님들이 아이들에게 사주는 상품, 정해서 오는 경우도 있지만 사전 정보가 다른 카테고리에 비해 부족할 수 있기에 사이트 내에서 얻는 정보도 중요하게 작용한다고 생각.
    - 사이트의 소비 가격 특징에서도 부합하고 수요도 많은 것이 장난감이라고 생각.

4. 사진 수가 늘어나면서 순위 변동이 생기지만 이것이 제품의 판매량 순위라기 보다는 제품 구매를 하고자 하는 소비자가 원래 구매할 목적으로 들어온 상품과 사이트 내에서 정보를 얻어 구매하는 상품으로 나눠지는 것 같음
    - 구매할 목적 : 사진이 적어도 판매량이 높고 사진수에 큰 영향을 받지 않음
    - 정보를 얻어 구매하는 목적 : 순위권내에는 존재하지만 사진이 많아질수록 순위가 올라오는 경향이 있음
    - *애초에 위 두 집단이 독립적인 집단이 아닌 기본적으로 상품을 구매할 목적은 있지만 정보를 얻음으로해서 구매의사가 바뀌는 상품이 있는 것 같다고 추정(장난감, 미용...)
    - *사이트 자체가 고가의 제품을 취급하기보다는 한국의 쿠팡과 같은 입지인 것 같고 저가의 제품에 대해서는 상세 설명이 제품 구매에 큰 영향을 안주는 거 같음
