````python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time

# 검색 키워드(자유롭게 작성)
search_keyword = "키보드"
url = f"https://search.shopping.naver.com/search/all?query={search_keyword}"

# 스크롤 다운 입력 사이 대기 시간
scroll_pause_time = 0.2
# 스크롤 시도 횟수
max_scroll_attempts = 20

# 페이지 이동 횟수
# next_count = range(1,101)


service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
driver.implicitly_wait(5)
driver.get(url)

actions = ActionChains(driver)
item_list = []
i = 0
while i < 4:
    for _ in range(max_scroll_attempts):
        actions.send_keys(Keys.PAGE_DOWN).perform()
        time.sleep(scroll_pause_time)

    # 정보 탐색

    product_list = driver.find_elements(By.CSS_SELECTOR, "div.product_item__MDtDF")
    for product in product_list:
        try:
            name = product.find_element(By.CSS_SELECTOR, "div.product_title__Mmw2K")
            # print(name.text)
            price = product.find_element(By.CSS_SELECTOR, "span.price")
            a_tag = product.find_element(By.TAG_NAME, "a")
            link = a_tag.get_attribute("href")
            detail = product.find_element(
                By.CSS_SELECTOR, "div.product_detail_box__QjZP8"
            )
            wireless = detail.find_element(
                By.CSS_SELECTOR, "a.product_detail__oWDMs.linkAnchor"
            )

            keyborad_info = {
                "상품명": name.text,
                "가격": price.text,
                "상품 정보 주소": link,
                "무선": wireless.text,
            }

            item_list.append(keyborad_info)
        except:
            pass
    try:
        next_button = driver.find_element(By.LINK_TEXT, "다음")
        next_button.click()
    except:
        print("마지막 페이지입니다")
        break
    finally:
        i += 1
    print(i)


import json

with open("Keyboard_info.json", "w") as file:
    json.dump(item_list, file, indent=4, ensure_ascii=False)

# 상품 정보 딕셔너리 저장할 리스트
# item_list = []

# 아래에 코드를 작성하세요.
````
