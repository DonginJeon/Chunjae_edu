```python
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")
titanic = sns.load_dataset("titanic")
```

```python
print(penguins)
```

### (1) 펭귄 데이터 시각화

1. 성별로 구분한 서식지(섬)별 몸무게 평균을 시각화하세요.
    
    ![Untitled](https://prod-files-secure.s3.us-west-2.amazonaws.com/f2678325-6f7b-4a25-b188-86c42030d6d5/abca8144-d47d-4684-8740-2c57c0d0e1e7/Untitled.png)

```python
sns.barplot(data=penguins, x="island", y="body_mass_g", hue="sex")

plt.show()
```

2. 섬으로 구분한 종의 수를 시각화하세요.
    
    ![Untitled](https://prod-files-secure.s3.us-west-2.amazonaws.com/f2678325-6f7b-4a25-b188-86c42030d6d5/4f47153d-2bab-4824-a7b9-68a35947576b/Untitled.png)

```python
sns.countplot(data=penguins, x="species", hue="island")

plt.show()
```

3. 펭귄 종 별 무게의 분포도를 Box Plot으로 시각화하세요.
    
    ![Untitled](https://prod-files-secure.s3.us-west-2.amazonaws.com/f2678325-6f7b-4a25-b188-86c42030d6d5/4e47b999-399e-438e-8e1c-d38d4c8b8c04/Untitled.png)

```python
sns.boxplot(data=penguins, y="body_mass_g", x="species")

plt.show()
```

4. 펭귄 종 별 무게의 분포도를 Violin Plot으로 시각화하세요.
    
    ![Untitled](https://prod-files-secure.s3.us-west-2.amazonaws.com/f2678325-6f7b-4a25-b188-86c42030d6d5/50f3dc51-5e29-45aa-a126-f5c7e6b136bf/Untitled.png)

```python
sns.violinplot(data=penguins, y="body_mass_g", x="species")

plt.show()
```

5. 성별로 구분한 펭귄 종 별 무게의 분포도를 Violin Plot으로 시각화하세요.
    
    ![Untitled](https://prod-files-secure.s3.us-west-2.amazonaws.com/f2678325-6f7b-4a25-b188-86c42030d6d5/5bd090bd-09ad-4f0d-973b-7292578552bb/Untitled.png)

```python
sns.violinplot(data=penguins, y="body_mass_g", x="species", hue="sex")

plt.show()
```

### (2) 타이타닉 데이터 시각화

```python
sns.barplot(
    data=titanic,
    x="pclass",
    y="age",
)

plt.show()
```
