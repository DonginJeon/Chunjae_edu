```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# pyplot을 활용한 line plot
# plt.plot()

# seaborn을 활용한 scatterplot
# scatterplot(data = dataframe,x=열1,y=열2)

sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
# 식사인원수,청구금액관계를 산점도로 그리기
sns.scatterplot(data=tips, x="size", y="total_bill")
plt.show()
```

```python
# hue 인자 : 데이터를 구분할 수 있다.

import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# pyplot을 활용한 line plot
# plt.plot()

# seaborn을 활용한 scatterplot
# scatterplot(data = dataframe,x=열1,y=열2)

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker")
plt.show()
```

```python
# 여러개의 axes에 seaborn 그래프 그리기
tips = sns.load_dataset("tips")
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

sns.scatterplot(data=tips, x="total_bill", y="tip", ax=ax1)
plt.show()
```

```python
# sns.lineplot()

import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("flights")

sns.lineplot(data=df, x="year", y="passengers")
plt.title("Basic Line Plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

g = sns.FacetGrid(
    data=tips,
    col="day",
    row= "time",  # dat열을 기준으로 그래프들을 시각화
)

g.map(plt.hist,"total_bill",)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

g = sns.FacetGrid(
    data = tips,
    col = "smoker",
    row = "time"
)

g.map(sns.histplot,"tip")
plt.show()
```
