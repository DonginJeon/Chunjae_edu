```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1,5,1,5,1]



plt.plot(x, y, label="y")
# 타이틀 설정
plt.title("Line Plot")
# 라벨 설정
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# 범례 표시
plt.legend()

# 그래프 출력
plt.show()
```

```python
# Figure,Axes 생성해서 그래프 시각화

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1,5,1,5,1]

## Figure와 Axes를 생성하는 함수
fig, ax = plt.subplots()

# 실제로 그래프가 그려지는 공간 : Axes
ax.bar(x,y,label = "x-y")
ax.set_xlabel("x axis")
ax.set_ylabel("y axis")
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1,5,1,5,1]


#nrow, ncol 인자로 생성할 axes 개수를 지정
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2, ncols =2)

# 첫 번째 (0,0) axes, line plot
ax1.plot(x,y)

# 두번째 axes
ax2.bar(x,y,label = "y")
ax2.set_xlabel("x axis")
ax2.set_ylabel("y axis")
ax2.set_title("bar")
ax2.legend()

#3 세번째 axes
ax3.scatter(x,y,label = "y")


fig.suptitle("1 figure, 4 axes")

#그래프 출력
plt.show()
```
