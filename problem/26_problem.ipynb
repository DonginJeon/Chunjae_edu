### (1) 지도에 마커 1개 그리기

1. Map 객체를 아래 인자로 생성합니다.
    - `location=[37.5665, 126.9780]`
    - `zoom_start=13`

```python
import folium
m= folium.Map(location=[37.5665,126.9780],zoom_start=13)
```

2. Marker 객체를 아래 위도 경도에 생성하고 마커를 지도에 추가합니다. 
    - 위도 : 37.47875057229362
    - 경도 : 126.88665978428628

```python
folium.Marker(
    location=[37.47875057229362,126.88665978428628]).add_to(m)
m
```

3. 적절한 `popup` `tooltip` `icon`을 설정하세요.

```python
name = "마리오 아울렛2관"
html = f"<p style='font-size:10px'>{name}</p>"
popup = folium.Popup(
    html=html,
    max_width=300,
    min_width=100
)
icon = folium.Icon(icon="pencil",color="green")

folium.Marker(
    location=[37.47875057229362,126.88665978428628],
    popup=popup,
    icon=icon,
    tooltip="클릭!"
    ).add_to(m)
m
```

```python
name = "공중화장실"
html = f"<p style='font-size:10px'>{name}</p>"
popup = folium.Popup(
    html=html,
    max_width=300,
    min_width=100
)
icon = folium.Icon(icon="flag",color="green")


pb_m = fl.Map(location=[37.47875057229362, 126.88665978428628],zoom_start=15)
    
for i in range(len(df['위도'])):
    a = df.loc[i, '위도']
    b = df.loc[i,'경도']
    folium.Marker(
    location=[a,b],
    name= name,
    html= html,
    popup=popup,
    icon=icon
    ).add_to(pb_m)
pb_m
```
