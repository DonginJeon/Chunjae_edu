```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")
print(diamonds)
car_crashes = sns.load_dataset("car_crashes")
print(car_crashes)
penguins = sns.load_dataset("penguins")
```

### (1) Heatmap

1. 다이아몬드 데이터 중 아래 열에 대해서 Heatmap을 시각화하세요.
    - `carat` `depth` `price`
    
    ![Untitled](https://prod-files-secure.s3.us-west-2.amazonaws.com/f2678325-6f7b-4a25-b188-86c42030d6d5/7ca970ed-9c7b-4a03-83e7-4aa7b8a4e1fd/Untitled.png)

```python
a = diamonds[["carat", "depth", "price"]]
dia_corr = a.corr()
sns.heatmap(dia_corr, annot=True)
```

2. 미국 주 별 자동차 사고 데이터 `car_crashes` 중 아래 열에 대해서 Heatmap을 시각화하세요.
    - `total` : 주당 100만 명당 총 사고 수
    - **`speeding`** : 주당 100만 명당 속도 위반 사고 수.
    - **`alcohol`** : 주당 100만 명당 음주 운전 사고 수.
    - **`not_distracted`** : 주당 100만 명당 비주의 분산 사고 수.
    - **`no_previous`** : 주당 100만 명당 이전에 사고를 낸 적 없는 운전자가 일으킨 사고 수.
    
    ![Untitled](https://prod-files-secure.s3.us-west-2.amazonaws.com/f2678325-6f7b-4a25-b188-86c42030d6d5/4043580f-3765-4609-b23f-2956efd781b1/Untitled.png)

```python
b = car_crashes[["total", "speeding", "alcohol", "not_distracted", "no_previous"]]
car_corr = b.corr()
sns.heatmap(car_corr, annot=True)
```

### (2) Pie Chart

1. 다이아몬드 데이터에서 cut의 비율을 Pie Chart로 시각화하세요.
    
    ![Untitled](https://prod-files-secure.s3.us-west-2.amazonaws.com/f2678325-6f7b-4a25-b188-86c42030d6d5/f040d2a5-ae81-4582-81f0-702b478020c6/Untitled.png)

```python
cut_u = diamonds["cut"].unique()
cut_v = diamonds["cut"].value_counts()
plt.pie(x=cut_v, labels=cut_u, autopct="%1.10f%%")
```

2. 펭귄 데이터에서 펭귄 종의 비율을 Pie Chart로 시각화하세요.
    
    ![Untitled](https://prod-files-secure.s3.us-west-2.amazonaws.com/f2678325-6f7b-4a25-b188-86c42030d6d5/f8786962-0b6e-4910-afda-58b6662b7c50/Untitled.png)

```python
pen_u = penguins["species"].unique()
pen_v = penguins["species"].value_counts()
plt.pie(x=pen_v, labels=pen_u, autopct="%1.2f%%")
```
