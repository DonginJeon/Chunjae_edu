```python
import spacy
from spacy import displacy
```

```python
import subprocess

# spacy ko_core_news_sm 모델 다운로드
subprocess.run(["python", "-m", "spacy", "download", "ko_core_news_sm"])
```

```python
nlp = spacy.load("ko_core_news_sm")
```

```python
doc = nlp("이것은 한국어로 된 문장입니다")
```

```python
for token in doc:
    print(token.text, token.dep_, token.head.text)
```

```python
displacy.render(doc, style="dep", jupyter=True)
```

```python
doc_1 = nlp(
    "두려움은 당신을 어두운 세계로 이끈다. 두려움은 분노를 낳고, 분노는 증오를 낳는다. 그리고 증오는 고통을 낳는다."
)
```

```python
for token in doc_1:
    print(token.text, token.dep_, token.head.text, token.lemma_, token.tag_)
```

```python
for token in doc_1:
    print(token.lemma_)
```

```python
displacy.render(doc_1, style="dep", jupyter=True)
```

```python
def syntictic():
    doc = nlp(input("분석을 원하는 문장을 입력해주세요"))
    return displacy.render(doc, style="dep", jupyter=True)
```

```python
syntictic()
```
