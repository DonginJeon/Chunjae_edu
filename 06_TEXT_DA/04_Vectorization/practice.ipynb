```python
# TF-IDF 벡터화
import numpy as np


vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(combined_list)

# TF-IDF 행렬을 배열로 변환
tfidf_array = tfidf_matrix.toarray()

# 단어 레이블
labels = vectorizer.get_feature_names_out()

# 각 문서에 대한 개별 바 플롯 생성 (상위 10개 단어만)
num_docs = tfidf_array.shape[0]

for i in range(num_docs):
    plt.figure(figsize=(12, 6))

    # 상위 10개 단어의 TF-IDF 점수 추출
    top_indices = np.argsort(tfidf_array[i])[-10:]  # 상위 10개 인덱스
    top_labels = labels[top_indices]
    top_scores = tfidf_array[i][top_indices]

    plt.bar(top_labels, top_scores)

    # 레이블과 타이틀 설정
    plt.xlabel("Words")
    plt.ylabel("TF-IDF Scores")
    plt.title(f"Top 10 TF-IDF Scores for Document {i + 1}")
    plt.xticks(rotation=45, ha="right")

    # 그래프 표시
    plt.tight_layout()
    plt.show()
```
