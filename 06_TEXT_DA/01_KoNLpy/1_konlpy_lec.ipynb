### konlpy 형태소 분석 (토큰화)
##### 형태소: 뜻을 가진 가장 작은 말의 단위
* 자립 형태소: 접사, 어미, 조사와 상관없이 자립하여 사용할 수 있는 형태소. 체언(명사, 대명사, 수사), 수식언(관형사, 부사), 감탄사 등
* 의존 형태소: 다른 형태소와 결합하여 사용되는 형태소. (접사, 어미, 조사, 어간)

* morphs : 형태소 추출
* pos : 품사 태깅
* nouns : 명사만 추출

```python
from konlpy.tag import Kkma, Hannanum, Komoran, Okt

hannanum = Hannanum()
kkma = Kkma()
komoran = Komoran()
# mecab = Mecab()
okt = Okt()
```

```python
# nsmc 데이터 로드
train_data = pd.read_csv('https://raw.githubusercontent.com/e9t/nsmc/master/ratings_train.txt', sep='\t', encoding='utf-8')
train_data.to_csv('./data/ratings_train.csv', index=False, encoding='utf-8')
```

```python
# N번째 문서 가져오기
sample_data2 = train_data.iloc[100]['document']
sample_data2
```

```python
# compare_morphs method 만들기 -> 한 번에 모든 morphs 를 볼 수 있게 출력해보기
def compare_morphs(sample_data):
    print(f'hannanum : {hannanum.morphs(sample_data)}')
    print(f'kkma : {kkma.morphs(sample_data)}' )
    print(f'komoran : {komoran.morphs(sample_data)}')
    print(f'okt : {okt.morphs(sample_data)}')
```

```python
# 실행
compare_morphs(sample_data2)
# 이름: 신 카이 마코토, 미유와 하나 카나
# 원형: Komoran
# 맞춤법 : 었다
# 기타: hannanum 작화와,미유 더대박이
```

품사 태깅 (Pos)

```python
# compare_pos method 만들기 -> 한 번에 모든 pos 를 볼 수 있게 출력해보기
def compare_pos(sample_data):
    print(f'hannanum : {hannanum.pos(sample_data)}')
    print(f'kkma : {kkma.pos(sample_data)}' )
    print(f'komoran : {komoran.pos(sample_data)}')
    print(f'okt : {okt.pos(sample_data)}')
```

```python
# 실행
compare_pos(sample_data2)
# 신카이 마코토: NNP (고유명사)
# EP: 선어말 어미
```

명사 추출 (nouns)

```python
# compare_nouns method 만들기 -> 한 번에 모든 명사를 볼 수 있게 출력해보기
def compare_nouns(sample_data):
    print(f'hannanum : {hannanum.nouns(sample_data)}')
    print(f'kkma : {kkma.nouns(sample_data)}' )
    print(f'komoran : {komoran.nouns(sample_data)}')
    print(f'okt : {okt.nouns(sample_data)}')
```

```python
compare_nouns(sample_data2)
```

### 품사, 명사 별 count 하기, 기초 시각화

```python
# 문서 1000개에 대해서 품사 별로 count 해보기
pos_dict = {}
for tag in okt.tagset.keys():
    pos_dict[tag] = 0

print(pos_dict)

for doc in train_data['document'][:1000]:
    doc_pos = okt.pos(str(doc))
    for uni_pos in doc_pos :
        pos = uni_pos[1] # (단어, 품사)
        if pos in pos_dict:
            pos_dict[pos] = pos_dict[pos] + 1
        else :
            pos_dict[pos] = 1

print(pos_dict)

# {'Adjective': 189508, 'Adverb': 44652, 'Alpha': 12954, 'Conjunction': 4135, 'Determiner': 17825, 'Eomi': 518, 'Exclamation': 6353, 'Foreign': 6605, 'Hashtag': 4, 'Josa': 360544, 'KoreanParticle': 27531, 'Noun': 896632, 'Number': 27229, 'PreEomi': 14, 'Punctuation': 204877, 'ScreenName': 5, 'Suffix': 46007, 'Unknown': 0, 'Verb': 266856, 'Modifier': 32968, 'VerbPrefix': 10894, 'URL': 48, 'Email': 9}
```

```python
# plot.bar 그려보기: values count / index 품사
pd.DataFrame(pos_dict.values(), index=pos_dict.keys(), columns=['counting']).plot.bar()
```
